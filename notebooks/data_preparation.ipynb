```python
import pandas as pd

df = pd.read_csv("1000_companies.csv")

df.to_excel("1000_companies.xlsx", index=False)
```

```python
from faker import Faker
```

```python
fake = Faker()

data = {
    "ID": [i for i in range(1, 1001)], 
    "Company Name": [fake.company() for _ in range(1000)]
}

df = pd.DataFrame(data)

df.to_csv("1000_companies_with_id.csv", index=False)


df.to_excel("1000_companies_with_id.xlsx", index=False)

df.head()
```

```python
df = pd.read_excel("1000_companies.xlsx")
```

```python
df["ID"] = range(1, len(df) + 1)

fake = Faker()
df["Company Name"] = [fake.company() for _ in range(len(df))]

df.to_excel("1000_companies_with_names.xlsx", index=False)

df.head()
```

```python
cols = ["ID", "Company Name"] + [col for col in df.columns if col not in ["ID", "Company Name"]]
df = df[cols]

df.to_excel("1000_companies_final.xlsx", index=False)

df.head()
```
